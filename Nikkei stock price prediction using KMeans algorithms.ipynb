```python
new_ndata2["Open_Close"] = new_ndata2.Close - new_ndata2.Open
new_ndata2["High_Low"] = new_ndata2.High - new_ndata2.Low
new_ndata2 = new_ndata2.dropna()
X = new_ndata2[["Open_Close", "High_Low"]]
X.head()
```

```python
X = X[["Open_Close", "High_Low"]]
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
inertia = []
for k in range(1, 16):
  kmeans = KMeans(n_clusters= k)
  kmeans.fit(scaled_data)
  centroid = kmeans.cluster_centers_
  inertia.append(kmeans.inertia_)


plt.figure(figsize= (8, 4))
plt.plot(range(1, 16), inertia, marker = "o")
plt.xlabel("Number of cluster")
plt.ylabel("INERTIA")
plt.title("Clustering in Nikkei data set")
plt.show()

kmeans = KMeans(n_clusters = 3, random_state= 42)
clusters = kmeans.fit_predict(scaled_data)
```
